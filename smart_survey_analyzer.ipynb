RASTGELE ANKET VERİSİ

```python
import random

scores = []

while len(scores) < 20:
    score = random.randint(0, 6)

    if score == 0:
        break

    if score == 6:
        continue  # Bu score listeye eklenmeden döngünün başına döner

    scores.append(score)
```

VERİ KONTROLÜ

```python
gecerli_puan = len(scores)
```

ORTALAMA HESAPLAMA

```python
toplam_puan = 0
if len(scores) > 0:
    for i in scores:
        toplam_puan +=i
        ortalama_puan = toplam_puan / len(scores)
```

MİN / MAX ANALİZİ

```python
en_dusuk  = min(scores)
en_yuksek = max(scores)
```

MEMNUNİYET ANALİZİ

```python
memnun = 0
kararsiz = 0
memnun_degil = 0

for puan in scores:
    if puan in (4, 5):
        memnun += 1
    elif puan == 3:
        kararsiz += 1
    elif puan in (1, 2):
        memnun_degil += 1

```

ERKEN UYARI SİSTEMİ

```python
bir = 0

for puan in scores:
    if puan ==1:
        bir += 1
        if bir == 3:
            print("Kritik memnuniyetsizlik tespit edildi")
            break
    else:
        bir = 0


```

GENEL ÖZET

```python
print(f'Toplam Geçerli Cevap: {gecerli_puan}\nOrtalama Puan: {ortalama_puan}\nMin Puan: {en_dusuk}\nMax Puan:{en_yuksek}\n\nMemnun: {memnun}\nKararsız: {kararsiz}\nMemnun Değil: {memnun_degil} ')
```
