HAM SİPARİŞ VERİSİ

```python
orders = [
    {
"customer":"Ahmet",
"items": ["Laptop","Mouse","Mouse"],
"total":18500,
"is_paid":True
    },
    {
"customer":"Zeynep",
"items": ["Keyboard"],
"total":2500,
"is_paid":False
    },
    {
"customer":"Mehmet",
"items": ["Monitor","HDMI Cable","Monitor"],
"total":7200,
"is_paid":True
    },
    {
"customer":"Mehmet",
"items": ["Monitor","HDMI Cable","Monitor"],
"total":7200,
"is_paid":True
    }
]
```

SİPARİŞ SAYISI ANALİZİ

```python
#Toplam sipariş sayısını bul
total_orders = len(orders)
```

```python
#Ödenmiş sipariş sayısını bul
paid_orders = 0

if orders[0]["is_paid"]:
    paid_orders = paid_orders + 1

if orders[1]["is_paid"]:
    paid_orders = paid_orders + 1

if orders[2]["is_paid"]:
    paid_orders = paid_orders + 1

if orders[3]["is_paid"]:
    paid_orders = paid_orders + 1
```

```python
#Ödenmemiş sipariş sayısını bul
unpaid_orders = total_orders - paid_orders
```

MÜŞTERİ LİSTESİ

```python
#Tüm müşteri isimlerini bir liste haline getir
customer_list = [
    orders[0]["customer"],
    orders[1]["customer"],
    orders[2]["customer"],
    orders[3]["customer"]
]
```

```python
#Tekrar eden müşteri varsa set kullanarak temizle
unique_customers = set(customer_list)
```

ÜRÜN ANALİZİ

```python
#Tüm siparişlerde geçen ürünleri bir listeye topla
all_items = (
    orders[0]["items"] +
    orders[1]["items"] +
    orders[2]["items"] +
    orders[3]["items"]
)
```

```python
#Unique ürünleri set ile ayır
unique_items = set(all_items)
```

SİPARİŞ DURUMU KONTROLÜ

```python
#Ahmet'in sipariş durumu
if orders[0]["is_paid"]:
    status_0 = "Sipariş onaylandı"
else:
    status_0 = "Ödeme bekleniyor"
```

```python
#Zeynep'in sipariş durumu
if orders[1]["is_paid"]:
    status_1 = "Sipariş onaylandı"
else:
    status_1 = "Ödeme bekleniyor"
```

```python
#Mehmet'in sipariş durumu (sipariş 3)
if orders[2]["is_paid"]:
    status_2 = "Sipariş onaylandı"
else:
    status_2 = "Ödeme bekleniyor"
```

```python
#Mehmet'in sipariş durumu (sipariş 4 — tekrar kayıt)
if orders[3]["is_paid"]:
    status_3 = "Sipariş onaylandı"
else:
    status_3 = "Ödeme bekleniyor"
```

GÜVENLİ SİPARİŞ ÖZETİ (Tuple)

```python
#Her sipariş için (müşteri_adı, toplam_tutar) tuple'ı oluştur
summary_0 = (orders[0]["customer"], orders[0]["total"])
summary_1 = (orders[1]["customer"], orders[1]["total"])
summary_2 = (orders[2]["customer"], orders[2]["total"])
summary_3 = (orders[3]["customer"], orders[3]["total"])
```

```python
#Tüm tuple özetlerini bir listeye ekle
order_summaries = [summary_0, summary_1, summary_2, summary_3]
```

GENEL ÇIKTILAR

```python
print("Toplam sipariş:", total_orders)
print("Ödenmiş sipariş:", paid_orders)
print("Ödenmemiş sipariş:", unpaid_orders)
print()
print("Unique ürünler:", unique_items)
print("Sipariş özetleri:", order_summaries)
```
